# training the model.

```python
from sklearn.model_selection import train_test_split

x = df.drop(['User_ID', 'Calories'], axis=1)
y = df['Calories'].values
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=22)
X_train.shape, X_test.shape
```

```python
# normalizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), Lasso(), RandomForestRegressor(), Ridge()]

for i in range(4):
    models[i].fit(X_train,y_train)
    print(f'{models[i]} : ')
    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(y_train,train_preds))
    val_preds = models[i].predict(X_test)
    print('Validation Error : ',  mae(y_test, val_preds))
    print()
```

RandomForestRegressor: This model shows signs of overfitting due to the large difference between training and validation errors. However, it still performs better on the validation set compared to the linear models, suggesting it captures the underlying patterns in the data more effectively.
